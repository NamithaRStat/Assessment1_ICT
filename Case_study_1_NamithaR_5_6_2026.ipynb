# **2.Weather Data Analyzer**
```
a. Analyze weather data (temperature, humidity, etc.) for a given region
b. Use variables and data types to store weather data
c. Implement conditional statements to determine weather conditions
(sunny, rainy, etc.)
d. Use functions to calculate average temperature and humidity, and print
results
```


````python
def main():
  temp =[23,30,28.5]
  humid = [67,78,69.8]
  status = True
  temp_u = int(input("Temperature in Degree celcius")) # temperature given by the user
  humid_u = int(input("Humidity in Percentage (%) without % symbol"))   # humidity given by the user
  while status:
    if (temp_u<= 18 and humid_u <= 10):
     print("Winter")
    elif(18< temp_u <= 35 and 10< humid_u<= 45):
     print("Rainy")
    elif(35< temp_u <= 55 and 45< humid_u<= 85):
     print("Summer")
    else:
     print("Please check your entires")
    status = False
    break
  temp.append(temp_u)
  humid.append(humid_u)
  avg_temp = sum(temp)/len(temp)
  avg_humid = sum(humid)/len(humid)
  print("Average temperature :",avg_temp,", &",end =" ")
  print("Average Humidity :",avg_humid)
main()
````
